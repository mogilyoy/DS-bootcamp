```python
import seaborn as sns
import pandas as pd


data = pd.read_csv('./csv/winners.csv')
data.head()
```

```python
data.info()
```

```python
sns.histplot(data=data.wins, binwidth=0.5 )
```

```python
sns.scatterplot(data=data, x=data.games, y=data.wins, hue=data.school)
```

```python
sns.heatmap(data=data.corr())
```

```python
sns.barplot(data=data, x=data.school, y=data.wins)
```
