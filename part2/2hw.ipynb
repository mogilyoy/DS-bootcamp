```python
import numpy as np

a = np.array([[1, -2, 4], [2, 0, -1]])
b = np.array([[5, 2, 3], [4, 6, 2]])
a, b
```

```python
print(a + b)
print(a - b)
```

```python
a = np.array([[1, 2], [2, -1], [3, 0]])
b = np.array([[-1 ,1], [1, 2], [0, 0]])
a, b
```

```python
c = a - 3*b
c
```

```python
a = np.array([[1, -1], [2, 0], [3, 0]])
b = np.array([[1, 1], [2, 0]])
a, b
```

```python
c = a@b
c
```

```python
a = np.array([[6, -1, 4], [2, 1, 10], [8, 0, 2]])
np.linalg.det(a)
```

```python
d = np.array([[3, 3, -1], [4, 1, 3], [1, -2, 2]], dtype='int64')
e = np.linalg.inv(d)
d, e, d@e
```
